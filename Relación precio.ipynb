```python
import pandas as pd
import matplotlib.pyplot as plt

#utilizamos el df_sin_outliers
df = pd.read_csv('https://raw.githubusercontent.com/JM181998/PROYECTOAIRBNB/refs/heads/main/df_alquileres_ordenado.csv')
```

```python
def scatter_segmentado(df, x, y, segmentar_por):
    ubicaciones = df[segmentar_por].unique()
    plt.figure(figsize=(10, 6))

    for ubicacion in ubicaciones:
        subset = df[df[segmentar_por] == ubicacion]
        plt.scatter(subset[x], subset[y], label = ubicacion, alpha=0.7)
    
    plt.title(f'Relación entre {y} y {x} segmentado {segmentar_por}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend(title=segmentar_por, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

scatter_segmentado(df, 'superficie_construida', 'precio', 'CCAA')


```

```python
#prueba para quitar outliers - no merece la pena


Q1 = df['precio'].quantile(0.25)
Q3 = df['precio'].quantile(0.75)
IQR = Q3 - Q1

print(f'Q1 = {Q1}')
print(f'Q3 = {Q3}')
print(f'IQR = {IQR}')

#definimos los límites para detectar outliers

lim_inferior = Q1 - 3 * IQR
lim_superior = Q1 + 3 * IQR

#filtrar el df para eliminar outliers

df = df[(df['precio'] >= lim_inferior) & (df['precio'] >= lim_superior)]

df.shape




```

```python
media = df['precio'].mean()
std = df['precio'].std()

# Filtrar registros dentro de ± 3 desviaciones estándar
df = df[(df['precio'] >= media - 3 * std) & (df['precio'] <= media + 3 * std)]
print(df.shape)
```
