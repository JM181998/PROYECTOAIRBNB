```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep

navegador = webdriver.Chrome()
navegador.get("https://www.pisos.com/alquiler_viviendas/")

# Aceptar cookies 
try:
    navegador.find_element(by=By.CSS_SELECTOR, value="#didomi-notice-disagree-button").click()
    sleep(1) 
except:
    pass 

elementos_ubicaciones = navegador.find_elements(by=By.CSS_SELECTOR, value="ul.seo-box__locations-items > li.seo-box__location-item > a.seo-box__location-link--level2")

urls_ubicaciones = []
for item in elementos_ubicaciones:
    nombre_ubicacion = item.text
    url_ubicacion = item.get_attribute('href')
    urls_ubicaciones.append(url_ubicacion)
    print(f"{nombre_ubicacion} - {url_ubicacion}")

paginas_a_scrapear = []

for url_ubicacion in urls_ubicaciones:
    nombre_ubicacion = url_ubicacion.split('/')[-2] 
    
    print(f"\nAnalizando la ubicación: {nombre_ubicacion}")
    numero_pagina = 1
    while True:
        url_pagina = f"{url_ubicacion}{numero_pagina}/" 
        
        # Navegar a la página
        navegador.get(url_pagina)
        sleep(2) 

        try:
            no_resultados = navegador.find_element(by=By.CSS_SELECTOR, value="div.no-results")
            if no_resultados.is_displayed():
                print(f"   Fin de las páginas de {nombre_ubicacion}")
                break 
        except:
            print(f"   La página {numero_pagina} de {nombre_ubicacion} existe")
            paginas_a_scrapear.append(url_pagina)
            numero_pagina += 1

    print(f"Fin del análisis de la provincia {nombre_ubicacion}")
    
navegador.quit()
```

```python
with open("paginas_a_scrapear.txt", "w") as archivo:
    for pagina in paginas_a_scrapear:
        archivo.write(pagina + "\n")
```
