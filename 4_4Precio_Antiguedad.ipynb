```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hola = "alquilereslimpio-checkpoint.csv"
df = pd.read_csv(hola)
```

```python
#en mi opinion respondiendo a la pregunta de punto 4 no existe una relacion clara entre precio y antiguedad, ademas que los años van en ragons de 10 o 20 años, pero parece ser
# que va la relacion del precio va mas en su ubicacion o tamaño
#de todos modos cuando nos reunamos lo miramos todos
```

```python
#creo un boxplot y obersvamos que hay varios outliers sin sentido ya que el df es de alquiler, tendremos que revisarlo y poceder a su eliminacion si corresponde
plt.figure(figsize=(10, 6))
sns.boxplot(x='antiguedad', y='precio', data=df)
plt.xlabel('Antigüedad (años)')
plt.ylabel('Precio (USD)')
plt.title('Distribución del Precio según la Antigüedad')
```

```python
# Agrupar por antigüedad y calcular el precio promedio, aqui obervamos mejor la diferencia y el analisis 
df_grouped = df.groupby('antiguedad')['precio'].mean().reset_index()

# Usar seaborn para crear un gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='antiguedad', y='precio', data=df_grouped, palette='viridis')
plt.xlabel('Antigüedad rango de años)')
plt.ylabel('Precio Promedio')
plt.title('Precio Promedio según la Antigüedad')
plt.show()
```

```python
#otro tipo de grafico similiar al anterior pero con mejor visualizacion

if 'antiguedad' in df.columns and 'precio' in df.columns:
    # Eliminar valores nulos
    df_filtered = df.dropna(subset=['antiguedad', 'precio'])
    
    # Contar la media de precio por categoría de antigüedad
    precio_promedio = df_filtered.groupby('antiguedad')['precio'].mean().sort_index()

    # Graficar
    plt.figure(figsize=(12, 6))
    precio_promedio.plot(kind='bar')
    plt.title('Relación entre Precio Promedio y Antigüedad')
    plt.xlabel('Antigüedad')
    plt.ylabel('Precio Promedio')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
else:
    print("Las columnas 'antiguedad' o 'precio' no existen en el DataFrame.")
```
